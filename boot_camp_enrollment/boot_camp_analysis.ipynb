```python
# Dependencies
import pandas as pd
```

```python
# PART I
# load CSV
closedfile = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
```

```python
# Read with pandas
df = pd.read_csv(closedfile)
df
```

```python
# Inspect all columns
df.columns
```

```python
# Extract only columns needed for analysis
df = df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
df.columns
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
df['AttendedBootcamp'] = df['AttendedBootcamp'].replace({0 : 'No', 1 : 'Yes'})
```

```python
# Calculate total number of respondents in survey
respondents = len(df)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attendees = df.loc[df['AttendedBootcamp'] == 'Yes']
attendees
```

```python
# Calculate average age of attendees
avg_age = attendees['Age'].mean()
avg_age
```

```python
# Calculate how many people attended a bootcamp
attendee_count = attendees['AttendedBootcamp'].count()
attendee_count
```

```python
# Calculate percentage of respondents who attended a bootcamp
bootcamp_pct = len(attendees) / len(df) 
bootcamp_pct
```

```python
# Calculate how many attendees hold degrees
attendees['SchoolDegree'].value_counts()
```

```python
# Count the number of records where the person is a degree holder
degree = attendees["SchoolDegree"].loc[
    (attendees["SchoolDegree"] == "bachelor's degree") |
    (attendees["SchoolDegree"] == "master's degree (non-professional)") |
    (attendees["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)") |
    (attendees["SchoolDegree"] == "associate's degree") |
    (attendees["SchoolDegree"] == "Ph.D.")
].count()
print(degree)
```

```python
# Determine the percentage of attendees with a degree
degree_pct = degree / attendee_count
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
gender = attendees["Gender"].value_counts()
gender
```

```python
# Calculate percentage of attendees belonging to each gender

total_gender = attendees['Gender']
males = attendees.loc[attendees['Gender'] == 'male']
females = attendees.loc[attendees['Gender'] == 'female']
nonbinary = attendees.loc[(attendees['Gender'] != 'male') & (attendees['Gender'] != 'female')]

male_pct = len(males)/len(total_gender)
female_pct = len(females)/len(total_gender) 
nonbinary_pct = len(nonbinary)/len(total_gender)

print(f'Male Percentage: {male_pct}, Female Percentage: {female_pct}, Nonbinary Percentage: {nonbinary_pct}')
```

```python
# Calculate average post-bootcamp salary
camp_salary = df['BootcampPostSalary'].mean()
camp_salary
```

```python
# Create a new table consolidating above calculations
campoverview_df = pd.DataFrame({"Total Respondents" : [respondents], "Total Bootcamp Attendees" : [attendee_count],  
                                "Respondent/Attendee %" : [bootcamp_pct], "Average Attendee Age" : [avg_age],
                                "% Degree" : [degree_pct], "% Male" : [male_pct], "% Female" : [female_pct], 
                                "% Other Genders" : [nonbinary_pct], "Avg. Post Camp Salary" : [camp_salary]})

campoverview_df
```

```python
# Improve formatting before outputting spreadsheet

campoverview_df['Total Respondents'] = campoverview_df['Total Respondents'].map("{:,}".format)
campoverview_df['Respondent/Attendee %'] = campoverview_df['Respondent/Attendee %'].map("{:.2%}".format)
campoverview_df['% Degree'] = campoverview_df['% Degree'].map("{:.2%}".format)
campoverview_df['% Male'] = campoverview_df['% Male'].map("{:.2%}".format)
campoverview_df['% Female'] = campoverview_df['% Female'].map("{:.2%}".format)
campoverview_df['% Other Genders'] = campoverview_df['% Other Genders'].map("{:.2%}".format)
campoverview_df['Average Attendee Age'] = campoverview_df['Average Attendee Age'].map("{:.2f}".format)
campoverview_df['Avg. Post Camp Salary'] = campoverview_df['Avg. Post Camp Salary'].map("${:.2f}".format)

campoverview_df
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
campoverview_df.to_excel('output/boot_camp_attendee_profile.xlsx', index = False)
```

```python
# PART II
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
campers = attendees.groupby(['BootcampName']).count()['AttendedBootcamp']
camper_df = pd.DataFrame({"Campers" : campers})
camper_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recs = attendees.groupby(['BootcampName']).sum()['BootcampRecommend']
recs_df = pd.DataFrame({"Recommendations": recs})
recs_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
boot_camps = pd.merge(camper_df, recs_df, on = ["BootcampName"])
boot_camps.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
boot_camps['% Recommended'] = boot_camps['Recommendations'] / boot_camps['Campers']

# Sort results in descending order
boot_camps = boot_camps.sort_values("% Recommended", ascending = False)

# Format for percentages
boot_camps['% Recommended'] = boot_camps['% Recommended'].map("{:.2%}".format)
boot_camps.head()
```

```python
# Export to excel and remove index
final = boot_camps.reset_index()
final.to_excel('output/boot_camp_program_profile.xlsx', index = False)
final
```
