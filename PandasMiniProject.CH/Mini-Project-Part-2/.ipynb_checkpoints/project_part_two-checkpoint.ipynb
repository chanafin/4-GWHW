```python
# Dependencies
import pandas as pd
```

```python
# load CSV
closedfile = "Resources/2016-FCC-New-Coders-Survey-Data.csv"

df = pd.read_csv(closedfile)
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file

df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc

df = df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]

df.columns
```

```python
# Change reading 0 and 1 to No and Yes, respectively

df['AttendedBootcamp'] = df['AttendedBootcamp'].replace({0 : 'No', 1 : 'Yes'})


df.head()
```

```python
# Extract rows for only those who attended a bootcamp

attendee = df.loc[df['AttendedBootcamp'] == 'Yes']

attendee
```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts

counts = attendee.groupby(['BootcampName']).count()
enrollees = counts['AttendedBootcamp']

enrollee_df = pd.DataFrame({"Enrollees" : enrollees})

enrollee_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

recs = attendee.groupby(['BootcampName']).sum()['BootcampRecommend']

recs_df = pd.DataFrame({"Recommendations": recs})

recs_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp

table = pd.merge(enrollee_df, recs_df, on = ["BootcampName"])
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator

table['% Recommended'] = table['Recommendations'] / table['Enrollees']

# Sort results in descending order
sorted_table = table.sort_values("% Recommended", ascending = False)

# Format for percentages
sorted_table['% Recommended'] = sorted_table['% Recommended'].map("{:.2%}".format)

sorted_table.head()
```

```python
# Export to excel and remove index
final = sorted_table.reset_index()

final
```

```python
final.to_excel('output/BootcampOutputPart2.xlsx', index = False)
```
